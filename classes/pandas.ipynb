```python
# Grouping and aggregation
df.groupby("country")["income"].mean()

df.groupby("country")["income"].sum()


# if we want to calculat mean for each gender then aggregate for gender

df.groupby("gender")["income"].mean()

# first group by country and then apply multiple aggregation function

df.groupby("country")["income"].agg(["mean","min","max"])
df.groupby("country")["income"].aggregate(["mean","min","max"])

df.groupby("country")["income"].agg(avg_salary = "mean",min_salary = "min",max_salary = "max")
```

```python
# can also group and aggreagate for multiple columns

df.groupby("country").agg({"income":"mean","age":"mean"})

df.groupby("country").agg({"income":"mean","age":"max"})

df.groupby("country").agg(max_salary = ("income","max"),avg_age = ("age","mean"))
```

```python
# melt and pivot

#first create some DataFrame
df = pd.DataFrame({
    "country":["USA","USA","INDIA","INDIA"],
    "years":[2020,2021,2020,2021],
    "sales":[100,120,90,110],
    "profit":[20,25,18,22],
    "tax":[10,20,30,40]
})

melted_data = df.melt(id_vars = ["country","years"],
        value_vars= ["sales","profit","tax"],
        var_name = "metrics",
        value_name = "value"
       )
melted_data

```

# Merging and joining data

```python
df_customers = pd.DataFrame({
    "customer_id":[1,2,3,4],
    "name":["Adam","Bob","Charlie","Dave"]
})

df_orders = pd.DataFrame({
    "order_id" :[101,102,103,104],
    "customer_id":[2,1,4,5],
    "amount":[250,120,300,180]
})

inner_joined = pd.merge(df_customers,df_orders,on = "customer_id",how = "inner")  #takes only intersectiong values of both the DataFrames
left_joined = pd.merge(df_customers,df_orders,on = "customer_id",how = "left")  #takes only intersectiong values of both the DataFrames
right_joined = pd.merge(df_customers,df_orders,on = "customer_id",how = "right")#takes only intersectiong values of both the DataFrames
outer_joined = pd.merge(df_customers,df_orders,on = "customer_id",how = "outer")  #takes only intersectiong values of both the DataFrames


```

```python
print(inner_joined,left_joined,right_joined,outer_joined)
```

```python
# concatenation
df1 = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Adam", "Eve", "Bob"]
})

df2 = pd.DataFrame({
    "id": [4, 5, 6],
    "name": ["Charlie", "Fester", "Gilian"]
})

pd.concat([df1,df2]) # row by row
pd.concat([df1,df2],ignore_index = True)

pd.concat([df1,df2],axis = 1)
```
