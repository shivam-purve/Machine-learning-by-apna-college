```python
# working on IRIS DATASET  showing their first 10 rows
# shape and data types and summary statistics
import pandas as pd
df = pd.read_csv("IRIS.csv")
df.head(10)
print(df.shape,df.dtypes)

df.describe()
```

```python
filtered_data = df.query(
    "(petal_length > 4.5) & (species == 'Iris-virginica')"
)
filtered_data
```

```python
df.groupby("species").agg(
    avg_sepal_len=("sepal_length", "mean"),
    max_petal_width=("petal_width", "max"),
    std_dev=("sepal_width", "std")
)
```

```python
df["petal_ratio"] = df["petal_length"] / df["petal_width"].replace(0, float("nan"))
df
```

```python
df.groupby("species")["petal_ratio"].agg(avg_petal_ratio = "mean")
```
