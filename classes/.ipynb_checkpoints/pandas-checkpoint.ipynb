# Filtering data

```python
# Filtering data based on different conditions


df[df["aqi"]>100]

df[(df["aqi"]>100) & (df["temperature"]>30)]

#Chaining
aqi_data = df[(df["aqi"]>100)&(df["temperature"]>30)][["city","aqi"]]
aqi_data

aqi_data.iloc[0]
#aqi_data.loc[0]  # doesnt gives anything as label starts from 6
aqi_data.loc[6]

```

```python
#Select on based of query -  by defalut return copy not view
df.query("aqi>150")
df.query("aqi>100 & temperature>30")[["city","aqi"]]

aqi_val = int(input("enter threshold for aqi: "))
temp_val=int(input("enter threshold for temperature: "))
aqi_data = df.query("aqi>@aqi_val & temperature>@temp_val")[["city","aqi"]].copy()
print(aqi_data)
```

# cleaning data

```python
df = pd.read_csv("raw_data.csv")
```

```python
df.isnull()

df.isna()
df.isnull().sum()
#df.dropna(axis = 0)

#replace all missing agee valus with its averae
cleaned_data = df.copy()
avg_mean = df["age"].mean()
cleaned_data["age"] = cleaned_data["age"].fillna(avg_mean)
cleaned_data = cleaned_data.drop_duplicates()

df.ffill()
df.bfill()
cleaned_data
```

```python
cleaned_data
```

```python
#Handl Duplicate values

df.duplicated()
df2 = df.copy()

df[["country","gender"]].duplicated()
df.drop_duplicates()
```
