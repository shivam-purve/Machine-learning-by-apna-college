```python
import requests
```

```python
URL = "https://www.scrapethissite.com/pages/simple/"
res = requests.get(URL)

```

```python
print(res.status_code)
```

```python
if res.status_code==200:
    print(res.headers)
```

```python
import os

os.makedirs("scrapped_data", exist_ok=True)

with open("scrapped_data/data1.html", "w", encoding="utf-8") as f:
    f.write(res.text)
```

```python
from bs4 import BeautifulSoup

with open("scrapped_data/data1.html","r") as f:
    html_content = f.read()
soup = BeautifulSoup(html_content,"lxml")
```

```python
soup.find("h3")
soup.h1
all_h3 = soup.find_all("h3")
```

```python
all_h3
type(all_h3)
```

```python
for h3 in all_h3:
    print(h3.get_text(strip = True))
```

```python
all_countries = []


for h3 in all_h3:
    name = h3.get_text(strip = True)
    # print(h3.find_parent("div")["class"])
    population = h3.find_next("div").select("span.country-population")[0].get_text(strip = True)
   # print(h3.find_next("div").select_one("span.country-population").get_text(strip = True))
    all_countries.append([name,population])
```

```python
all_countries
```

```python
import pandas as pd
df = pd.DataFrame(all_countries)
```

```python
df.columns = ["country_name","population"]
```

```python
df
```

```python
import os
os.makedirs("cleaned_data")
df.to_csv("cleaned_data/country-pop.csv",index = False)
```
